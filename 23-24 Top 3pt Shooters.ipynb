```python
import pandas as pd
import numpy as np

stats = pd.read_csv("23_NBA_Stats.csv")
```

```python
stats
```

```python
# Select the desired columns
useful_columns = [
    'PLAYER',
    'TEAM',
    'AGE',
    'GP',
    'MIN',
    '3PA',
    '3P%'
]

shooters = stats[useful_columns].copy()
```

```python
shooters
```

```python
shooters['Shooting Score'] = round((shooters['3PA'] * 0.7) + (shooters['3P%'] * 0.85) + (shooters['GP'] * 0.4), 2)
shooters.sort_values(by='Shooting Score', ascending=False)
```

```python
def shooter_type(shooters):

    # Lasers
    if (shooters['3PA'] >= 7) & (shooters['3P%'] >= 40):
        return 'High Volume Laser'
    elif (4 <= shooters['3PA'] < 7) & (shooters['3P%'] >= 40):
        return 'Mid Volume Laser'
    elif (2 <= shooters['3PA'] < 4) & (shooters['3P%'] >= 40):
        return 'Low Volume Laser'
    elif (shooters['3PA'] < 2) & (shooters['3P%'] >= 40):
        return 'Hesitant Laser'
    # Respectable Shooters
    elif (shooters['3PA'] >= 7) & ((35 <= shooters['3P%']) & (shooters['3P%'] < 40)):
        return 'High Volume Respectable Shooter'
    elif (4 <= shooters['3PA'] < 7) & ((35 <= shooters['3P%']) & (shooters['3P%'] < 40)):
        return 'Mid Volume Respectable Shooter'
    elif (2 <= shooters['3PA'] < 4) & ((35 <= shooters['3P%']) & (shooters['3P%'] < 40)):
        return 'Low Volume Respectable Shooter'
    elif (shooters['3PA'] < 2) & ((35 <= shooters['3P%']) & (shooters['3P%'] < 40)):
        return 'Hesitant Respectable Shooter'
    # Capable Shooters
    elif (shooters['3PA'] >= 7) & ((30 <= shooters['3P%']) & (shooters['3P%'] < 35)):
        return 'High Volume Capable Shooter'
    elif (4 <= shooters['3PA'] < 7) & ((30 <= shooters['3P%']) & (shooters['3P%'] < 35)):
        return 'Mid Volume Capable Shooter'
    elif (2 <= shooters['3PA'] < 4) & ((30 <= shooters['3P%']) & (shooters['3P%'] < 35)):
        return 'Low Volume Capable Shooter'
    elif (shooters['3PA'] < 2) & ((30 <= shooters['3P%']) & (shooters['3P%'] < 35)):
        return 'Hesitant Capable Shooter'
    # Self-check Shooters
    else:
        return 'Self-Check'
    
shooters['Shooter Type'] = shooters.apply(shooter_type, axis=1)
```

```python
shooters
```

```python
shooters = shooters[shooters['GP']>=10]
shooters = shooters[shooters['3PA']>=2]
shooters = shooters.sort_values(by='Shooting Score', ascending=False)
shooters
```

```python
shooters['League Rank'] = shooters['Shooting Score'].rank(method='min',ascending=False)
shooters['Team Rank'] = shooters.groupby('TEAM')['Shooting Score'].rank(ascending=False)
```

```python
new_order = [
    'League Rank',
    'Team Rank',
    'PLAYER',
    'TEAM',
    'AGE',
    'GP',
    'MIN',
    '3PA',
    '3P%',
    'Shooting Score',
    'Shooter Type'
]

shooters = shooters[new_order].copy()
shooters
```

```python
prev_stats = pd.read_csv("22_NBA_Stats.csv")
prev_stats
```

```python
useful_columns = [
    'PLAYER',
    'TEAM',
    'AGE',
    'GP',
    'MIN',
    '3PA',
    '3P%'
]

prev_stats = prev_stats[useful_columns].copy()

prev_stats['Previous Shooting Score'] = round((prev_stats['3PA'] * 0.7) + (prev_stats['3P%'] * 0.85) \
                                     + (prev_stats['GP'] * 0.4), 2)

prev_stats = prev_stats.sort_values(by='Previous Shooting Score', ascending=False)
prev_stats
```

```python
prev_stats['Previous Shooter Type'] = prev_stats.apply(shooter_type, axis=1)
prev_stats
```

```python
with pd.ExcelWriter('prev_shooter_workbook.xlsx', engine='xlsxwriter') as writer:
    prev_stats.to_excel(writer, sheet_name='Sheet1', index=False)

import subprocess
import platform

file_path = 'prev_shooter_workbook.xlsx'

# Check the platform to determine the appropriate command
if platform.system() == 'Windows':
    command = ['start', 'excel', file_path]
elif platform.system() == 'Linux':
    command = ['xdg-open', file_path]
elif platform.system() == 'Darwin':  # macOS
    command = ['open', file_path]
else:
    raise Exception("Unsupported operating system")

# Use subprocess to open the file
subprocess.run(command, check=True)
```

```python
merged_df = pd.merge(
    shooters,
    prev_stats,
    left_on=['PLAYER'],
    right_on=['PLAYER'],
    how='left'
)

merged_df
```

```python
useful_columns = [
    'League Rank',
    'Team Rank',
    'PLAYER',
    'TEAM_x',
    'AGE_x',
    'GP_x',
    'MIN_x',
    '3PA_x',
    '3P%_x',
    'Shooting Score',
    'Shooter Type',
    'Previous Shooter Type'
]

merged_df = merged_df[useful_columns].copy()
merged_df
```

```python
merged_df['TEAM'] = merged_df['TEAM_x']
merged_df['AGE'] = merged_df['AGE_x']
merged_df['GP'] = merged_df['GP_x']
merged_df['MIN'] = merged_df['MIN_x']
merged_df['3PA'] = merged_df['3PA_x']
merged_df['3P%'] = merged_df['3P%_x']
```

```python
def shooter_reputation(merged_df):
    previous_shooter_type = str(merged_df['Previous Shooter Type'])
    shooter_type = str(merged_df['Shooter Type'])
    
    if 'Laser' in previous_shooter_type and 'Respectable' in shooter_type:
        return 'Regressed Laser'
    elif 'Laser' in previous_shooter_type and 'Capable' in shooter_type:
        return 'Regressed Respectable Shooter'
    elif 'Laser' in previous_shooter_type and shooter_type == 'Self-Check':
        return 'Regressed Capable Shooter'
    elif 'Respectable' in previous_shooter_type and 'Capable' in shooter_type:
        return 'Regressed Respectable Shooter'
    elif 'Respectable' in previous_shooter_type and shooter_type == 'Self-Check':
        return 'Regressed Capable Shooter'
    elif 'Capable' in previous_shooter_type and shooter_type == 'Self-Check':
        return 'Regressed Capable Shooter'
    else:
        return shooter_type

merged_df['Shooter Reputation'] = merged_df.apply(shooter_reputation, axis=1)
```

```python
new_order = [
    'League Rank',
    'Team Rank',
    'PLAYER',
    'TEAM',
    'AGE',
    'GP',
    'MIN',
    '3PA',
    '3P%',
    'Shooting Score',
    'Shooter Type',
    'Previous Shooter Type',
    'Shooter Reputation'
]

merged_df = merged_df[new_order].copy()
```
